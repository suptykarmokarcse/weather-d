This notebook contains tasks to calculate mean, median, mode, variance, and standard deviation for selected numerical columns of your dataset.

```python

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration for plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Load your dataset
# Using the uploaded file: 'weatherHistory (1).csv'
df = pd.read_csv('weatherHistory (1).csv')
print("First 5 rows of the dataset:")
df.head()

# Identify numerical columns for analysis
# Based on the dataset, we'll select 'Temperature (C)' and 'Wind Speed (km/h)'
num_cols = ['Temperature (C)', 'Wind Speed (km/h)']
print("\nNumerical Columns Selected:", num_cols)

# Data Cleaning (if necessary - handling missing values for selected columns)
# For simplicity, we'll drop rows with any missing values in the selected columns for accurate calculation.
df.dropna(subset=num_cols, inplace=True)
print(f"\nDataset size after dropping NaNs: {len(df)}")
```

## Task 1: Measures of Central Tendency
Select at least two numerical columns and calculate mean, median, and mode.

**Justification:**i have selected Temperature and Wind Speed.
**Temperature(C)**: This is a core variable in the dataset.Additionally it does represents the weather condition deeply and by analyzing the central tendency of this column will provide proper idea over typical temperature experiences.
**Wind Speed:**This column can give me the measurement of air condition and central tendency helps to understand typical windiness




```python
num_cols = ['Temperature (C)', 'Wind Speed (km/h)'] # Correct num_cols for this cell's execution
print("\n" + "="*50)
print("Task 1: Measures of Central Tendency")
print("="*50)

for col in num_cols:
    print(f"\n--- Analysis for: {col} ---")

    # Calculate Mean
    mean_val = df[col].mean()

    # Calculate Median
    median_val = df[col].median()

    # Calculate Mode
    # Mode can return multiple values, we take the first one for comparison or print all.
    mode_val = df[col].mode()

    print(f'Mean (x̄): {mean_val:.4f}')
    print(f'Median: {median_val:.4f}')

    # Check if there is a unique mode or multiple modes
    if len(mode_val) == 1:
        print(f'Mode: {mode_val.iloc[0]:.4f}')
    else:
        # In continuous data, the mode is often not meaningful or has many duplicates.
        # We can display the top few most frequent values or the mode from the bins if using a histogram approach.
        print(f'Mode (most frequent value(s)): {mode_val.tolist()}')

    # Interpretation: Compare Mean, Median, and Mode to assess skewness
    if mean_val > median_val:
        print("Interpretation: Mean > Median, suggesting a slight positive (right) skewness.")
    elif mean_val < median_val:
        print("Interpretation: Mean < Median, suggesting a slight negative (left) skewness.")
    else:
        print("Interpretation: Mean ≈ Median, suggesting a relatively symmetric distribution.")
```

## Task 2: Measures of Dispersion
Calculate variance and standard deviation for the selected columns.

```python
num_cols = ['Temperature (C)', 'Wind Speed (km/h)'] # Ensure correct columns for this cell
print("\n" + "="*50)
print("Task 2: Measures of Dispersion")
print("="*50)

dispersion_results = {}

for col in num_cols:
    print(f"\n--- Analysis for: {col} ---")

    # Calculate Variance (s^2)
    variance_val = df[col].var()

    # Calculate Standard Deviation (s)
    std_dev_val = df[col].std()

    print(f'Variance (s²): {variance_val:.4f}')
    print(f'Standard Deviation (s): {std_dev_val:.4f}')

    dispersion_results[col] = {'std': std_dev_val}

# Interpretation of Dispersion
print("\n--- Comparative Interpretation ---")
std_temp = dispersion_results['Temperature (C)']['std']
std_wind = dispersion_results['Wind Speed (km/h)']['std']

print(f"Standard Deviation of Temperature (C): {std_temp:.2f}")
print(f"Standard Deviation of Wind Speed (km/h): {std_wind:.2f}")

if std_temp < std_wind:
    print(f"Conclusion: 'Temperature (C)' (Std Dev: {std_temp:.2f}) is relatively more consistent (less variable) than 'Wind Speed (km/h)' (Std Dev: {std_wind:.2f}).")
    print("The smaller standard deviation for temperature means the hourly temperature readings tend to cluster closer to the mean temperature than the wind speed readings cluster around the mean wind speed.")
else:
    print(f"Conclusion: 'Wind Speed (km/h)' (Std Dev: {std_wind:.2f}) is relatively more consistent (less variable) than 'Temperature (C)' (Std Dev: {std_temp:.2f}).")
    print("The smaller standard deviation for wind speed means the hourly wind speed readings tend to cluster closer to the mean wind speed than the temperature readings cluster around the mean temperature.")
```

## Task 3: Visualization
Plot histograms with mean, median, and mode indicated.

```python
# Plot histograms with mean, median, and mode indicated.
print("\n" + "="*50)
print("Task 3: Visualization (Histograms)")
print("="*50)

for col in num_cols:
    plt.figure(figsize=(10,6))

    # Calculate statistics for plotting lines
    col_mean = df[col].mean()
    col_median = df[col].median()
    col_mode = df[col].mode().iloc[0] if len(df[col].mode()) > 0 else None

    # Create the histogram
    sns.histplot(df[col], bins=30, kde=False, color='#4A90E2', edgecolor='black', alpha=0.7)

    # Overlay lines for central tendency
    plt.axvline(col_mean, color='red', linestyle='solid', linewidth=2, label=f'Mean: {col_mean:.2f}')
    plt.axvline(col_median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {col_median:.2f}')
    if col_mode is not None:
        plt.axvline(col_mode, color='orange', linestyle='dotted', linewidth=2, label=f'Mode: {col_mode:.2f}')

    plt.title(f'Histogram of {col} with Central Tendency Measures', fontsize=16)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Frequency (Count)', fontsize=12)
    plt.legend()
    plt.show()

    # Comment on how dispersion is visible in the histogram:
    print(f"\n--- Visual Dispersion Comment for {col} ---")
    print("Dispersion (Variability) is visible by how spread out the data bars are.")
    print("A wider, flatter distribution indicates a higher standard deviation (more variability), while a tall, narrow distribution indicates a lower standard deviation (more consistency/less variability).")
    print(f"For '{col}', observe the total range of the x-axis and the spread around the central lines (mean/median). This visual confirms the standard deviation calculated in Task 2.")
```

Interpret the results of your calculations and visualizations. Discuss central tendency, spread, and any interesting observations from the data.
